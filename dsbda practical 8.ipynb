```python
import matplotlib.pyplot as plt
import seaborn as sns
data = sns.load_dataset('tips')
data.head()
```

```python
xyz = data[['total_bill','tip','size']]
xyz.corr()
```

```python
# 1. Dist Plot
sns.displot(data["total_bill"], kde=True, color="red")
plt.title("Distribution of Total Bill")
plt.show()
```

```python
# 2. Joint Plot
jp = sns.jointplot(x="total_bill", y="tip", data=data, kind="scatter", color="purple")
jp.fig.suptitle("Joint Plot : Total Bill vs Tip")
jp.fig.subplots_adjust(top=0.90)
plt.show()
```

```python
# 3. Rug Plot
sns.rugplot(data=data, x="total_bill", height=0.1)
plt.title("Rug Plot of Total Bill")
plt.show()
```

```python
# 4. Bar Plot
sns.barplot(x="sex", y="total_bill", data=data, hue="sex", palette="pastel", legend=False)
plt.title("Average Total Bill by Sex")
plt.show()
```

```python
# 5. Count Plot
sns.countplot(x="day", data=data, palette="Set2", hue = "sex", legend = False)
plt.title("Count of Records by Day")
plt.show()

```

```python
# 6. Box Plot
sns.boxplot(x="day", y="total_bill", data=data, palette="Set3", hue = "sex", legend = False)
plt.title("Box Plot of Total Bill by Day")
plt.show()
```

```python
# 7. Violin Plot
sns.violinplot(x="day", y="total_bill", data=data, palette="muted",hue = "sex", legend = False)
plt.title("Violin Plot of Total Bill by Day")
plt.show()
```

```python
# 8. Strip Plot
sns.stripplot(x="day", y="total_bill", data=data, jitter=True, palette="cool", hue = "sex", legend = False)
plt.title("Strip Plot of Total Bill by Day")
plt.show()
```

```python
# 9. Swarm Plot
sns.swarmplot(x="day", y="total_bill", data=data, palette="rocket", hue = "sex", legend = False)
plt.title("Swarm Plot of Total Bill by Day")
plt.show()
```

```python
# 10. Matrix Plot (Pairplot)
pp = sns.pairplot(data, hue="sex", palette="husl")
pp.fig.suptitle("Pairplot of Tips Dataset")
pp.fig.subplots_adjust(top=0.90)
plt.show()
```

```python
# 11. Heat Map
corr = xyz.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Heatmap of Correlations")
plt.show()
```

```python
# 12. Distribution Plot (More than one variable)
sns.displot(data=data, x="total_bill", hue="sex", kde=True, palette="Set1")
plt.title("Distribution of Total Bill by Sex")
plt.show()
```

```python
# 13. Cluster Map
sns.clustermap(corr, annot=True)
plt.suptitle("Cluster Map of Correlations")
plt.show()
```
